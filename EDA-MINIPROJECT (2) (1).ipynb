```python
c=[]
for i in fifa.Value:
    if i[-1]=='M':
            m = float(i[1:-1])
            f = m*1000000
            c.append(f)
    elif i[-1]=='K':
            k = float(i[1:-1])
            j = k*1000
            c.append(j)
    else:
        j=float(i[1:])
        c.append(j)

m=np.array(c)
fifa.Value=m
```

```python
c=[]
for i in fifa.Wage:
    if i[-1]=='M':
            m = float(i[1:-1])
            f = m*1000000
            c.append(f)
    elif i[-1]=='K':
            k = float(i[1:-1])
            j = k*1000
            c.append(j)
    else:
        j=float(i[1:])
        c.append(j)

m=np.array(c)
fifa.Wage=m
```

```python
from datetime import datetime
from dateutil.parser import parse
import pandas as pd
```

```python
fifa['Joined'] = pd.DatetimeIndex(fifa['Joined']).year  
```

```python
p=[]
for i in fifa.Joined:
    if isinstance(i,float)!=True:
        v = int(i[-4:])
        p.append(v)
        
    else:
        p.append(i)

m = np.array(p)
fifa.Joined=m
```

```python
fifa.head()
```

```python
y =[]
import datetime
for i in fifa['Contract Valid Until']:
    if isinstance(i,float)==False:
        if len(i) == 4:
            d = datetime.datetime.strptime(i, '%Y')
            y.append(d)
        else:
            d = datetime.datetime.strptime(i, '%b %d, %Y')
            y.append(d)
    else:
        y.append(i)
            
o = np.array(y)
fifa['Contract Valid Until']=o
```

```python
fifa.head()
```

```python
j=[]
for i in fifa.Height:
    if isinstance(i,float)==False:
        o = i.replace("'",".")
        j.append(float(o))
    else:
        j.append(i)
m=np.array(j)
fifa.Height=m
```

```python
p=[]
for i in fifa.Weight:
    if isinstance(i,float)!=True:
        v = float(i[:3])
        p.append(v)
        
    else:
        p.append(i)

m = np.array(p)
fifa.Weight=m
```

```python
fifa.head()
```

```python
c=[]
for i in fifa['Release Clause']:
    if isinstance(i,float)==False:
        if i[-1]=='M':
            m = float(i[1:-1])
            f = m*1000000
            c.append(f)
        elif i[-1]=='K':
            k = float(i[1:-1])
            j = k*1000
            c.append(j)
        else:
            j=float(i[1:])
            c.append(j)
    else:
        c.append(i)
m=np.array(c)
fifa['Release Clause']=m
```
